```python
# Import required libraries and dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaled_df = scaler.fit_transform(market_data_df) 
```

```python
# SAMPLE DATA
# 	price_change_percentage_24h	price_change_percentage_7d	price_change_percentage_14d	price_change_percentage_30d	price_change_percentage_60d	price_change_percentage_200d	price_change_percentage_1y
# coin_id							
# bitcoin	0.508529	0.493193	0.772200	0.235460	-0.067495	-0.355953	-0.251637
# ethereum	0.185446	0.934445	0.558692	-0.054341	-0.273483	-0.115759	-0.199352
# tether	0.021774	-0.706337	-0.021680	-0.061030	0.008005	-0.550247	-0.282061
# ripple	-0.040764	-0.810928	0.249458	-0.050388	-0.373164	-0.458259	-0.295546
# bitcoin-cash	1.193036	2.000959	1.760610	0.545842	-0.291203	-0.499848	-0.270317

# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_df, columns=market_data_df.columns)

# Copy the crypto names from the original data
scaled_df["coin_id"] = market_data_df.index

# # Set the coin_id column as index
scaled_df = scaled_df.set_index("coin_id")

# Display sample data
scaled_df.head()
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# SAMPLE DATA
# k	inertia
# 0	1	287.000000
# 1	2	212.123342
# 2	3	145.897940
# 3	4	131.457370
# 4	5	66.317106
# 5	6	57.402668
# 6	7	49.212644
# 7	8	44.799804
# 8	9	33.859468
# 9	10	29.250314

# Create a list with the number of k-values to try
# Use a range from 1 to 11
k = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i, random_state=1)
    model.fit(scaled_df)
    inertia.append(model.inertia_)


# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k, 
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)


# Display the DataFrame
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

df_elbow.plot.line(x="k",
                   xlabel="Number of Clusters",
                   y="inertia",
                   ylabel="Inertia",
                   title="Elbow Curve",
                   xticks=k)
```

```python
# Determine the rate of decrease between each k value. 
k = elbow_data["k"]
inertia = elbow_data["inertia"]
for i in range(1, len(k)):
    percentage_decrease = (inertia[i-1] - inertia[i]) / inertia[i-1] * 100
    print(f"Percentage decrease from k={k[i-1]} to k={k[i]}: {percentage_decrease:.2f}%")
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:** The optimal value is k=4 because the most significant decrease occurs between k=3 and k=4 (52.15%). This is reinforced by the significant elbow in the graph. 

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, n_init='auto', random_state=1)

```

```python
# Fit the K-Means model using the scaled data
model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predictions = model.predict(scaled_df)


# View the resulting array of cluster values.
predictions
```

```python
# Create a copy of the DataFrame
scaled_df_predictions = scaled_df.copy()
```

```python
# SAMPLE DATA
# price_change_percentage_24h	price_change_percentage_7d	price_change_percentage_14d	price_change_percentage_30d	price_change_percentage_60d	price_change_percentage_200d	price_change_percentage_1y	crypto_cluster
# coin_id								
# bitcoin	0.508529	0.493193	0.772200	0.235460	-0.067495	-0.355953	-0.251637	1
# ethereum	0.185446	0.934445	0.558692	-0.054341	-0.273483	-0.115759	-0.199352	1
# tether	0.021774	-0.706337	-0.021680	-0.061030	0.008005	-0.550247	-0.282061	2
# ripple	-0.040764	-0.810928	0.249458	-0.050388	-0.373164	-0.458259	-0.295546	2
# bitcoin-cash	1.193036	2.000959	1.760610	0.545842	-0.291203	-0.499848	-0.270317	1

# Add a new column to the DataFrame with the predicted clusters
scaled_df_predictions["crypto_cluster"] = predictions

# Display sample data
scaled_df_predictions.head()
```

```python
# Create a scatter plot using Pandas plot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Use "rainbow" for the color to better visualize the data.
scaled_df_predictions.plot.scatter(x="price_change_percentage_24h",
                                   y="price_change_percentage_7d",
                                   c="crypto_cluster",
                                   colormap="rainbow",
                                   title="Crypto Clusters")
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# SAMPLE DATA
# array([[-0.60066733,  0.84276006,  0.46159457],
#        [-0.45826071,  0.45846566,  0.95287678],
#        [-0.43306981, -0.16812638, -0.64175193],
#        [-0.47183495, -0.22266008, -0.47905316],
#        [-1.15779997,  2.04120919,  1.85971527]])
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
crypto_pca = pca.fit_transform(scaled_df)


# View the first five rows of the DataFrame. 
crypto_pca[:5]
```

```python
# SAMPLE DATA
# array([0.3719856 , 0.34700813, 0.17603793])
# Retrieve the explained variance to determine how much information  can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

explained_variance
```

```python
# Sum the explained variance to determine how much information is retained by the three principal components.
explained_variance.sum()
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is 89.5%, which suggests that using these three principal components provides a strong approximation of the original data while reducing dimensionality.

```python
# SAMPLE DATA
# 	PCA1	PCA2	PCA3
# coin_id			
# bitcoin	-0.600667	0.842760	0.461595
# ethereum	-0.458261	0.458466	0.952877
# tether	-0.433070	-0.168126	-0.641752
# ripple	-0.471835	-0.222660	-0.479053
# bitcoin-cash	-1.157800	2.041209	1.859715
# binancecoin	-0.516534	1.388377	0.804071
# chainlink	-0.450711	0.517699	2.846143
# cardano	-0.345600	0.729439	1.478013
# litecoin	-0.649468	0.432165	0.600303
# bitcoin-cash-sv	-0.759014	-0.201200	-0.217653

# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
pca_df = pd.DataFrame(
    data=crypto_pca, columns=["PCA1", "PCA2", "PCA3"]
)

# Copy the crypto names from the original data
pca_df["coin_id"] = market_data_df.index


# Set the coin_id column as index
pca_df = pca_df.set_index("coin_id")


# Display sample data
pca_df.head(10)
```

```python
# Determine the rate of decrease between each k value. 
k = elbow_data["k"]
inertia = elbow_data["inertia"]
for i in range(1, len(k)):
    percentage_decrease = (inertia[i-1] - inertia[i]) / inertia[i-1] * 100
    print(f"Percentage decrease from k={k[i-1]} to k={k[i]}: {percentage_decrease:.2f}%")
```
